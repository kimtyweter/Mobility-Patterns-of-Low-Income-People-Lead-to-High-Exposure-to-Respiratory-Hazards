```python
IncomeGroup = 'Q1'
dfq1 = df_sec_final[df_sec_final['IncomeGroup'] == IncomeGroup]

nai_list = dfq1['naics code'].unique().tolist()
pos = []
neg = []
for i in range(0, len(nai_list)):
    val = dfq1[dfq1['naics code'] == nai_list[i]]['Delta_qr_log'].tolist()[0]
    
    if val >= 0:
        pos.append(val)
        neg.append(0)
    else:
        pos.append(0)
        neg.append(val)
nai_list = [str(x) for x in nai_list]

# --- plot---
from matplotlib.pyplot import figure
figure(figsize=(3, 10))

ax = plt.subplot(111)
ax.barh(nai_list, neg, color='b', height = 0.8, edgecolor='black')
ax.barh(nai_list, pos, color='r', height = 0.8, edgecolor='black')



ax.set_yticks(nai_list)
plt.plot([],[],color='r', label='+δ', linewidth=5)
plt.plot([],[],color='b', label='-δ', linewidth=5)

for spine in plt.gca().spines.values():
    spine.set_visible(True)
    spine.set_color('black')
    spine.set_linewidth('1')
    
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.setp(ax.spines.values(), linewidth=3)

plt.xticks(fontsize= 28)
plt.yticks(fontsize= 28)

plt.legend(fontsize = 24, framealpha=0, loc='upper left',bbox_to_anchor=[0.16, 1], bbox_transform=plt.gcf().transFigure)
plt.grid(False)



# plt.axvline(x=0, color='k', linestyle='--')

# plt.locator_params(axis = "y",nbins=5)
# ax = plt.gca()
plt.locator_params(axis = "x",nbins=4)
ax.set_xlim([-2, 6])


plt.subplots_adjust(bottom=0.15)
plt.subplots_adjust(left=0.15)

plt.tick_params(bottom = True)
plt.tick_params(left = True)

plt.xlabel("δ", fontsize= 28)
plt.ylabel("Sector", fontsize= 28)
plt.savefig('Sectors and activities in Q1.pdf', bbox_inches="tight")
```
